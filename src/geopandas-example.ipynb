```python
import pandas as pandas
import geopandas
import matplotlib
import xlrd
iller = geopandas.read_file("zip://data/iller.zip")
df = pandas.read_excel('./data/veriler.xlsx')
iller2 = iller.merge(df, on = "NAME_1")

ax_yillik_gelir = iller2.plot(figsize=(30,30), column='kisi basina yillik gelir (usd)', scheme='quantiles', legend=True)
ax_yillik_gelir.set_axis_off()
ax_yillik_gelir.set_title("Türkiye'de kişi başına düşen yıllık gelir görseli")

ax_nufus_dagilimi = iller2.plot(figsize=(30,30), column='nufus', scheme='quantiles', legend=True)
ax_nufus_dagilimi.set_axis_off()
ax_nufus_dagilimi.set_title("Türkiye'de nüfus dağılımı görseli")
```
